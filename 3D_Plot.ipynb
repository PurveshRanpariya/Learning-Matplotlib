```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = [10, 15, 12, 8, 5, 18, 7, 13, 9, 14]
y = [80, 75, 90, 85, 70, 95, 65, 78, 88, 82]
z = [20, 25, 30, 35, 40, 28, 32, 22, 38, 26]

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot 3D scatter
ax.scatter(x, y, z, c='blue', marker='^')   # blue triangles

# Labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title("3D Scatter Plot Example 2")

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  

# -----------------------------
# Generate sample data
# -----------------------------
z = np.linspace(0, 10, 100)   # Create 100 points between 0 and 10 for Z-axis
x = np.sin(z)                 # X values are sine of z (wave pattern)
y = np.cos(z)                 # Y values are cosine of z (wave pattern)

# -----------------------------
# Create 3D plot
# -----------------------------
fig = plt.figure(figsize=(8, 6))                      # Create figure with given size
ax = fig.add_subplot(111, projection='3d')            # Add a 3D subplot

# -----------------------------
# Plot 3D Line
# -----------------------------
ax.plot(x, y, z, color="green", linewidth=2)          # Draws a 3D line connecting (x, y, z)

# -----------------------------
# Customize labels & title
# -----------------------------
ax.set_xlabel("X-axis (sin values)")                  # Label for X-axis
ax.set_ylabel("Y-axis (cos values)")                  # Label for Y-axis
ax.set_zlabel("Z-axis (linear range)")                # Label for Z-axis
ax.set_title("3D Line Plot Example")                  # Plot title

# -----------------------------
# Show plot
# -----------------------------
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  

# Generate grid data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)                 
Z = np.sin(np.sqrt(X**2 + Y**2))         # Surface function

# Create 3D surface plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap="viridis")

# Add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Surface Plot")

plt.show()

```
